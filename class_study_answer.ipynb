```python
## 1
class Student:
    def __init__(self,name, scores=None):
        self.name = name
        self.scores = scores if scores is not None else {}
    def add_score(self, subject, score):
        self.scores[subject] = score
    
    def get_average(self):
        if not self.scores:
            return 0
        else:
            return sum(self.scores.values()) / len(self.scores)
    
    def print_info(self):
        print(f"학생 이름: {self.name}")
        for subject, score in self.scores.items():
            print(f"{subject}: {score}")
        print(f"평균 점수: {self.get_average():.2f}")

# 사용 예시
student_a = Student("철수", {"국어": 90, "수학": 80})
student_a.add_score("영어", 85)
student_a.print_info()
# 학생 이름: 철수
# 국어: 90
# 수학: 80
# 영어: 85
# 평균 점수: 85.00

##### 포인트: 학생마다 다른 점수를 저장해야 하므로 인스턴스 변수를 활용합니다.
##### 추가 확장: add_score에서 잘못된 점수(예: 음수)를 입력받지 못하도록 예외 처리 등을 적용할 수 있습니다.
```

```python
#2
import math

class Shape:
    def get_area(self):
        raise NotImplementedError("get_area 메서드를 서브클래스에서 구현하세요.")

class Rectangle(Shape):
    def __init__(self,width, height):
        self.width = width
        self.height = height
    
    def get_area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def get_area(self):
        return math.pi * (self.radius**2)

shapes=[Rectangle(4,5), Circle(3)]

for shape in shapes:
    area = shape.get_area()
    print("도형의 넓이 :" , area)
```

```python
#3
class Employee:
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary
    def get_annual_salary(self):
        return self.base_salary

    
class SalesEmployee(Employee):
    def __init__(self, name, base_salary, commision_rate):
        super().__init__(name, base_salary)
        self.commision_rate = commision_rate
    
    def get_annual_salary(self):
        return self.base_salary + (self.base_salary * self.commision_rate)
    
class EngineerEmployee(Employee):
    def __init__(self, name, base_salary, bonus):
        super().__init__(name, base_salary)
        self.bonus = bonus
    
    def get_annual_salary(self):
        return self.base_salary+self.bonus

# 사용 예시
emp1 = Employee("일반직원", 3000)
emp2 = SalesEmployee("영업사원", 3000, 0.2)
emp3 = EngineerEmployee("엔지니어", 3000, 500)

employees = [emp1, emp2, emp3]

for emp in employees:
    print(f"{emp.name} 연봉: {emp.get_annual_salary()}")
```

```python
#4
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

    def __str__(self):
        return f"<Book title={self.title}, author={self.author}, price={self.price}>"

class DiscountBook(Book):
    def __init__(self, title, author, price, discount_rate):
        super().__init__(title, author, price)
        self.discount_rate = discount_rate  # 예: 0.2 -> 20% 할인

    @property
    def price(self):
        # price를 조회할 때마다 할인율이 적용된 값을 반환
        # (상위 클래스에 정의된 self._price 사용 가정)
        discounted_price = self._price * (1 - self.discount_rate)
        return round(discounted_price, 2)

    @price.setter
    def price(self, new_price):
        # 세터: 상위 클래스의 _price에 직접 저장
        self._price = new_price

    def __str__(self):
        return f"<DiscountBook title={self.title}, author={self.author}, price={self.price}, discount={self.discount_rate}>"

class Bookstore:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def search_title(self, keyword):
        """keyword가 책 제목에 포함된 책들을 리스트로 반환"""
        result = []
        for book in self.books:
            if keyword.lower() in book.title.lower():
                result.append(book)
        return result

    def remove_book(self, title):
        """해당 제목의 책을 한 권만 제거"""
        for i, book in enumerate(self.books):
            if book.title == title:
                del self.books[i]
                return True
        return False

    def get_total_value(self):
        """보유 중인 모든 책의 가격 합계를 반환"""
        total = 0
        for book in self.books:
            total += book.price
        return total


# 사용 예시
store = Bookstore()
book1 = Book("Python Basic", "Alice", 30000)
book2 = Book("Advanced Python", "Bob", 35000)
book3 = DiscountBook("Data Science with Python", "Charlie", 40000, 0.15)

store.add_book(book1)
store.add_book(book2)
store.add_book(book3)

# 검색 테스트
result = store.search_title("python")
for b in result:
    print(b)
# <Book title=Python Basic, author=Alice, price=30000>
# <Book title=Advanced Python, author=Bob, price=35000>
# <DiscountBook title=Data Science with Python, author=Charlie, price=34000.0, discount=0.15>

# 총 가치
print("서점 총 책 가치:", store.get_total_value())  # 할인 적용된 가격이 포함되어야 함

# 책 제거 후 확인
store.remove_book("Python Basic")
print("서점 총 책 가치(제거 후):", store.get_total_value())


##### 포인트:
# DiscountBook 클래스에서 **오버라이딩 + 프로퍼티(@property)**를 사용해 할인된 가격을 동적으로 계산.
# Bookstore는 여러 책을 관리하므로, 내부적으로 list 자료구조를 사용해 Book 인스턴스들을 저장합니다.
# 검색, 삭제, 총합 계산 등의 메서드로 클래스-객체 간 협력을 연습해볼 수 있습니다.
```
